```python
##Plotting graph 
# Given values
R = 7.5  # Resistance in ohms
L = 3.3e-3  # Inductance in henries (3.3 mH)
frequencies = np.linspace(0, 20000, 500)  # Frequency range from 0 to 10kHz

# Calculate impedance magnitude for each frequency
omega = 2 * np.pi * frequencies  # Angular frequency
Z = np.sqrt(R**2 + (frequencies * L)**2)  # Magnitude of impedance

#Calculate phase
phi = np.arctan((frequencies*L)/R)

# Plot the relationship
plt.figure(figsize=(8, 6))
plt.plot(frequencies, Z, label='Impedance |Z|', color='orange')
plt.plot(frequencies, phi, label='Phase |phi|', color='blue')
plt.title('Magnitude of Impedance & Phase vs Frequency')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Impedance Magnitude (Ohms)/Phase (rads)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from scipy import signal
from scipy.fft import fft, fftfreq

def sine_function(t, amplitude, frequency, phase, offset):
    """Single sine function"""
    return amplitude * np.sin(2 * np.pi * frequency * t + phase) + offset

# Read the data
file_path = 'Lab3_sample_117.2Hz.csv'
data = pd.read_csv(file_path, skiprows=1)

# Extract time and accelerometer data
time = data['Time (s)'].values
accelerometer = data['Accelerometer (V)'].values

# Remove any NaN values
valid_indices = ~np.isnan(accelerometer)
time = time[valid_indices]
accelerometer = accelerometer[valid_indices]

# Center the data around zero
accelerometer_centered = accelerometer - np.mean(accelerometer)

# Calculate sampling frequency and design a narrower bandpass filter
fs = 1 / (time[1] - time[0])
nyquist = fs/2
center_freq = 117.2
bandwidth = 10  # Slightly wider bandwidth
low = (center_freq - bandwidth/2) / nyquist
high = (center_freq + bandwidth/2) / nyquist

# Use a lower order filter to prevent instability
filter_order = 4
b, a = signal.butter(filter_order, [low, high], btype='band')

# Apply zero-phase filtering with padding to prevent edge effects
padded_data = np.pad(accelerometer_centered, (100, 100), mode='edge')
filtered_padded = signal.filtfilt(b, a, padded_data)
accelerometer_filtered = filtered_padded[100:-100]  # Remove padding

# Check for any remaining NaN or inf values
if np.any(~np.isfinite(accelerometer_filtered)):
    print("Warning: Filtering produced invalid values. Applying alternative filtering method...")
    # Use a simpler filtering approach if the bandpass filter fails
    accelerometer_filtered = signal.medfilt(accelerometer_centered, kernel_size=3)

# Get initial amplitude from filtered data
initial_amplitude = np.std(accelerometer_filtered) * np.sqrt(2)

# Initial parameter guesses
p0 = [
    initial_amplitude,  # Amplitude
    117.2,             # Frequency (using known value)
    0,                 # Phase
    0                  # Offset (since we centered the data)
]

try:
    # Fit the sine function to the filtered data
    popt, pcov = curve_fit(sine_function, time, accelerometer_filtered, 
                          p0=p0, 
                          bounds=([0, 116, -2*np.pi, -0.1],      # Lower bounds
                                 [0.01, 118, 2*np.pi, 0.1]))     # Upper bounds
    
    # Generate points for the fitted curve
    time_fit = np.linspace(min(time), max(time), 1000)
    accelerometer_fit = sine_function(time_fit, *popt)
    
    # Calculate R-squared value
    fitted_points = sine_function(time, *popt)
    residuals = accelerometer_filtered - fitted_points
    ss_res = np.sum(residuals ** 2)
    ss_tot = np.sum(accelerometer_filtered ** 2)
    r_squared = 1 - (ss_res / ss_tot)
    
    # Plot results
    plt.figure(figsize=(15, 10))
    
    # Original data subplot
    plt.subplot(3, 1, 1)
    plt.plot(time, accelerometer, '.', label='Original Data', markersize=1, alpha=0.3)
    plt.ylabel('Accelerometer (V)')
    plt.title('Original Data')
    plt.grid(True)
    plt.legend()
    
    # Filtered data subplot
    plt.subplot(3, 1, 2)
    plt.plot(time, accelerometer_filtered, '.', label='Filtered Data', markersize=1, alpha=0.3)
    plt.plot(time_fit, accelerometer_fit, 'r-', label='Fitted Curve', linewidth=2)
    plt.ylabel('Accelerometer (V)')
    plt.title(f'Filtered Data with Fitted Curve (R = {r_squared:.6f})')
    plt.grid(True)
    plt.legend()
    
    # Residuals subplot
    plt.subplot(3, 1, 3)
    plt.plot(time, residuals, '.', markersize=1)
    plt.xlabel('Time (s)')
    plt.ylabel('Residuals (V)')
    plt.title('Fit Residuals')
    plt.grid(True)
    
    # Print fitted parameters
    print("\nFitted Parameters:")
    print(f"Amplitude: {abs(popt[0]):.6f} V")
    print(f"Frequency: {popt[1]:.2f} Hz")
    print(f"Phase: {popt[2]:.6f} rad")
    print(f"Offset: {popt[3]:.6f} V")
    print(f"R-squared: {r_squared:.6f}")
    
except RuntimeError as e:
    print(f"Error fitting the curve: {e}")
except ValueError as e:
    print(f"Value error during fitting: {e}")

plt.tight_layout()
plt.show()
```

```python
# Define a sine wave function for fitting
def sine_wave(t, A, f, phi, C):
    return A * np.sin(2 * np.pi * f * t + phi) + C

# Define a function to calculate R^2
def r_squared(y, y_fit):
    ss_res = np.sum((y - y_fit) ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    return 1 - (ss_res / ss_tot)

# Store the signals in a dictionary for easy iteration
signals = {
    'Load Cell': load_cell,
    'Accelerometer': accelerometer,
    'LED Voltage': led_voltage,
    'Coil Voltage': coil_voltage,
    'Coil Current': coil_current
}

# Initial guess for parameters: Amplitude, Frequency, Phase, Offset
initial_guess = [0.1, 117.2, 0, np.mean(load_cell)]
fit_params = {}
fitted_signals = {}
r2_scores = {}

# Fit each signal with a sine wave
for label, signal in signals.items():
    params, _ = curve_fit(sine_wave, time, signal, p0=initial_guess, maxfev=5000)
    fit_params[label] = params
    fitted_signal = sine_wave(time, *params)
    fitted_signals[label] = fitted_signal  # Store fitted values
    r2_scores[label] = r_squared(signal, fitted_signal)  # Calculate R^2

# Plotting results with R^2 displayed
fig, axs = plt.subplots(len(signals), 1, figsize=(10, 12))
fig.suptitle("Sine Wave Fitting for Each Signal with R^2 Scores")

for i, (label, signal) in enumerate(signals.items()):
    axs[i].plot(time, signal, 'b-', label='Original Data')
    axs[i].plot(time, fitted_signals[label], 'r--', label='Sine Fit')
    axs[i].set_xlabel("Time (s)")
    axs[i].set_ylabel(f"{label} (V or A)")
    axs[i].legend()
    axs[i].grid()
    # Display R^2 score on the plot
    axs[i].text(0.05, 0.95, f"$R^2$ = {r2_scores[label]:.4f}", transform=axs[i].transAxes, 
                verticalalignment='top', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Display the fitted parameters and R^2 scores
for label, params in fit_params.items():
    print(f"{label} - Amplitude: {params[0]:.4f}, Frequency: {params[1]:.4f} Hz, Phase: {params[2]:.4f}, Offset: {params[3]:.4f}, R^2: {r2_scores[label]:.4f}")
```
