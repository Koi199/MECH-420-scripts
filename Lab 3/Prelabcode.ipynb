```python
## Import libraries
# Data Manipulation
import pandas as pd  # For data manipulation and analysis
import numpy as np   # For numerical operations

# Data Visualization
import matplotlib.pyplot as plt  # For creating static, interactive, and animated visualizations
import seaborn as sns            # For statistical data visualization
from scipy.optimize import curve_fit
```

```python
##Plotting graph 
# Given values
R = 7.5  # Resistance in ohms
L = 3.3e-3  # Inductance in henries (3.3 mH)
frequencies = np.linspace(0, 20000, 500)  # Frequency range from 0 to 10kHz

# Calculate impedance magnitude for each frequency
omega = 2 * np.pi * frequencies  # Angular frequency
Z = np.sqrt(R**2 + (frequencies * L)**2)  # Magnitude of impedance

#Calculate phase
phi = np.arctan((frequencies*L)/R)

# Plot the relationship
plt.figure(figsize=(8, 6))
plt.plot(frequencies, Z, label='Impedance |Z|', color='orange')
plt.plot(frequencies, phi, label='Phase |phi|', color='blue')
plt.title('Magnitude of Impedance & Phase vs Frequency')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Impedance Magnitude (Ohms)/Phase (rads)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import least_squares

# Define the new sine model function
def sine_model(x, a, b, c, d):
    return a + b * np.sin(c * x + d)

# Define the residual function for least squares
def residuals(params, x, y):
    a, b, c, d = params
    return y - sine_model(x, a, b, c, d)

# Load your dataset (make sure to adjust the path or filename)
# dataset = pd.read_csv('your_file.csv')  # Uncomment if reading from a file

# Example data creation for demonstration (remove when using actual dataset)
# dataset = pd.DataFrame({
#     'Time (s)': np.linspace(0, 10, 100),
#     'Accelerometer (V)': 0.5 * np.sin(1.5 * np.linspace(0, 10, 100)) + np.random.normal(0, 0.1, 100)
# })

# Check if dataset is loaded properly
print("Dataset preview:")
print(dataset.head())

# Initial guess for the sine curve parameters: [offset, amplitude, frequency, phase shift]
initial_guess = [2.5, 1, 250, 1]

# Perform least squares optimization
result = least_squares(residuals, initial_guess, args=(dataset['Time (s)'], dataset['Coil Current (A)']))

# Get the fitted parameters from the result
a, b, c, d = result.x

# Formatted to 2 decimal places
a_formatted = f'{a:.2f}'
b_formatted = f'{b:.4f}'
c_formatted = f'{c:.4f}'
d_formatted = f'{d:.2f}'

# Calculate the fitted values using the original x-values
y_fit = sine_model(dataset['Time (s)'], a, b, c, d)

# Check the fitted values
print("Fitted values:")
print(y_fit[:5])  # Print first 5 fitted values for inspection

# Calculate R^2 score manually
def r2_score(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred) ** 2)
    ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)
    return 1 - (ss_res / ss_tot)

r2 = r2_score(dataset['Coil Current (A)'], y_fit)
r2_formatted = f'{r2:.2f}'

# Print fitted parameters and R^2 score
print(f"Fitted parameters: a = {a}, b = {b}, c = {c}, d = {d}")
print(f"R^2 score: {r2_formatted}")

frequencies = np.linspace(0, 2, 1000)  # Frequency range from 0 to 10kHz
#Calculate phase
phi = 2.544 + 0.002*np.sin(frequencies*250 + 1.57)


# Create the figure and axis
fig, ax = plt.subplots()

# Plot the data points and fitted curve
ax.plot(dataset['Time (s)'], dataset['Coil Current (A)'], marker='o', label='Data Points')
# ax.plot(frequencies, phi, marker='o', label='Data Points', markersize = 1)
ax.plot(dataset['Time (s)'], y_fit, label=rf'Fitted Line: $y = {a_formatted} + {b_formatted} \sin({c_formatted}x + {d_formatted})$', color='red')

# Move the x-axis and y-axis to intersect at zero
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Hide the top and right spines
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Set labels and title
ax.set_xlabel('Time (s)')
ax.set_ylabel('Coil Current (A)')
ax.set_title(rf'Curve Fitting: $y = a + b \sin(cx + d)$')

# Show the grid
ax.grid(True)

# Add the R^2 score as text annotation
ax.text(0.05, 0.9, f'$R^2 = {r2_formatted}$', transform=ax.transAxes, fontsize=12, verticalalignment='top')

# Display the legend
ax.legend(loc='lower right')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# # Load the dataset from the CSV file
# data = pd.read_csv('Lab3_sample_117.2Hz.csv')

# Assuming the CSV has columns named 'Time (s)' and 'Accelerometer (V)'
time = data['Time (s)'].values  # Extract time data
acceleration = data['Accelerometer (V)'].values  # Extract accelerometer data

# Stretch the x-axis for better visualization
stretch_factor = 10  # Adjust this factor as needed
time_stretched = time * stretch_factor

# Define the sine model function
def sine_model(x, a, b, c, d):
    return a + b * np.sin(c * x + d)

# Initial guess for the parameters: [offset, amplitude, frequency, phase shift]
initial_guess = [0, 1, 250, 0]

# Perform curve fitting using curve_fit
params, covariance = curve_fit(sine_model, time_stretched, acceleration, p0=initial_guess)

# Extract the fitted parameters
a, b, c, d = params

# Generate fitted values
y_fit = sine_model(time_stretched, a, b, c, d)

# Calculate R^2 score
ss_res = np.sum((acceleration - y_fit) ** 2)
ss_tot = np.sum((acceleration - np.mean(acceleration)) ** 2)
r2 = 1 - (ss_res / ss_tot)

# Print fitted parameters and R^2 score
print(f'Fitted parameters:')
print(f'a = {a:.2f}, b = {b:.2f}, c = {c:.2f}, d = {d:.2f}')
print(f'R^2 score: {r2:.2f}')

# Plot the data points and fitted curve
plt.figure()
plt.plot(time_stretched, acceleration, 'o', label='Data Points')
plt.plot(time_stretched, y_fit, 'r-', label=f'Fitted Line: y = {a:.2f} + {b:.2f} sin({c:.2f}x + {d:.2f})')
plt.xlabel('Time (s) (Stretched)')
plt.ylabel('Accelerometer (V)')
plt.title('Curve Fitting: y = a + b sin(cx + d) (Stretched X-axis)')
plt.grid(True)
plt.legend()
plt.show()
```
