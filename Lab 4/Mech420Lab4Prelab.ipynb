```python
import matplotlib.pyplot as plt
import numpy as np

# Given parameters
B_l = 10.2  # N/Amp
k = 1.4711  # N/mm

# Current values (in Amps)
current = np.linspace(0, 4, 100)

# Displacement calculation (in mm)
displacement = (B_l / k) * current

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(current, displacement, label=r'$y = \frac{10.2}{1.4711} \cdot x$', color='b')
plt.xlabel('Coil Current (A)')
plt.ylabel('Displacement (mm) ($x_{VCA}$)')
plt.title('Displacement vs Current (x = (10.2/1.4711)I)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Constants
k = 1.4711/(10**-3)  # Example value for k
m = 0.25  # Example value for m

# Frequency range for w
f = np.linspace(1, 1000, 1000)

# Calculate Z while avoiding division by zero
Z = abs(1/(k-((2*np.pi*f)**2 * m)))
# np.where(k - w**2 * m != 0, w / (k - w**2 * m), np.nan)

# Plotting
plt.figure(figsize=(8, 6))
plt.semilogx(f, Z, label='Impedance Magnitude (dB)', color='red')
plt.xlabel(r'Frequency f (Hz)')
plt.ylabel(r'$|Z(f)|$')
plt.title(r'Plot of $Z = \frac{1}{k - \omega^2 m}$ with respect to f (Hz) where $\omega$ = 2$\pi$f')
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
# plt.ylim(-10, 10)  # Adjust this range based on your values for k and m
plt.legend()
plt.grid(True)
plt.show()

print(np.sqrt(k/m))
```
