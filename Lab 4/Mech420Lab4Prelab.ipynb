```python
import matplotlib.pyplot as plt
import numpy as np

# Given parameters
B_l = 10.2  # N/Amp
k = 1.4711  # N/mm

# Current values (in Amps)
current = np.linspace(0, 4, 100)

# Displacement calculation (in mm)
displacement = (B_l / k) * current

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(current, displacement, label=r'$y = \frac{10.2}{1.4711} \cdot x$', color='b')
plt.xlabel('Coil Current (A)')
plt.ylabel('Displacement (mm) ($x_{VCA}$)')
plt.title('Displacement vs Current (x = (10.2/1.4711)I)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the parameters
k = 1471.1  # Stiffness (N/m)
m = 0.25   # Mass (kg)
R = 7.5
Bl = 10.2
L = 3.3/1000

# Calculate the frequency range
f = np.logspace(0, 3, 1000)  # Frequency range from 1 Hz to 1000 Hz
omega = 2 * np.pi * f        # Angular frequency

# Calculate the impedance magnitude
Z_total = np.sqrt((R**2) + ((omega*(L + (Bl**2)/(k - (omega**2)*m))))**2)

impedance_magnitude_db = 20 * np.log10(np.abs(Z_total))

# Plot the impedance magnitude in dB
plt.semilogx(f, impedance_magnitude_db)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Impedance Magnitude (dB)')
plt.title('Plot of Impedance Magnitude (dB) with respect to Frequency (Hz)')
plt.grid()
plt.show()

print((1/(2*np.pi))*np.sqrt(k/m))
```
